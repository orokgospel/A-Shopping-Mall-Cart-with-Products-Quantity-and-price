```python
input_data = """
Anne: Metallica, The_Doors, Black_Sabbath
John: The_Beatles, The_Doors, Metallica, Pink_Floyd
Kathy: U2, Guns_n_Roses, Led_Zeppelin
Jamie: Radiohead
Ashok: Guns_n_Roses, U2, Pink_Floyd, The_Doors
Sara: Blink_182, Iron_Maiden, The_Doors
"""

from collections import defaultdict, Counter

# Initialize data structures
band_counter = Counter()  # To count popularity
band_to_colleagues = defaultdict(list)  # To track colleagues for each band

# Step 1: Parse input
for line in input_data.strip().split('\n'):
    if not line.strip():
        continue
    # Split colleague name and bands
    colleague, bands_str = line.split(':', 1)
    bands = [b.strip() for b in bands_str.split(',') if b.strip()]

    # Update counts and colleague mapping
    for band in bands:
        band_counter[band] += 1
        band_to_colleagues[band].append(colleague.strip())

# Step 2: Top 2 liked bands (handle ties)
top_counts = band_counter.most_common()
if not top_counts:
    print("No bands found!")
else:
    max_count = top_counts[0][1]
    second_count = None

    # Find second place count
    for band, count in top_counts:
        if count < max_count:
            second_count = count
            break

    # Collect all bands that match 1st or 2nd count (including ties)
    top_bands = [band for band, count in band_counter.items()
                 if count == max_count or count == second_count]

    print("Top Bands:")
    for band in top_bands:
        print(band)

# Step 3: Band -> Colleagues mapping
print("\nBand to Colleagues mapping:")
for band, colleagues in band_to_colleagues.items():
    print(f"{band}: {', '.join(colleagues)}")
```
